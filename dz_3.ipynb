Построить одномерную линейную регрессию Direct_Bilirubin и Total_Bilirubin, предварительно разделив данные на 80% тренировочных и 20% тестовых. Построить график: исходных тренировочных и тестовых данных, а также прямой линейной регрессии.

```python
from sklearn.model_selection import train_test_split

dir_blrb = np.array(data['Direct_Bilirubin']).reshape(-1,1)
ttl_blrb = np.array(data['Total_Bilirubin']).reshape(-1,1)

x_train, x_test, y_train, y_test = train_test_split(dir_blrb, ttl_blrb, test_size = 0.2)
x_tt = x_test.reshape(1,-1)
y_tt = y_test.reshape(1,-1)
xx_tt = pd.Series(x_tt[0])
yy_tt = pd.Series(y_tt[0])

x_tr = x_train.reshape(1,-1)
y_tr = y_train.reshape(1,-1)
xx_tr = pd.Series(x_tr[0])
yy_tr = pd.Series(y_tr[0])

sns.regplot(x = xx_tt, y = yy_tt, fit_reg=True).set(title = 'Линейная регрессия по Direct_Bilirubin и Total_Bilirubin', xlabel = 'Direct_Bilirubin', ylabel = 'Total_Bilirubin' )
sns.regplot(x = xx_tr, y = yy_tr, fit_reg=True)
plt.xlabel('Direct_Bilirubin')
plt.ylabel('Total_Bilirubin')
plt.legend(['Тестовая выборка', 'Тренировочная выборка'])
```

```python
#рассчитаем линейную регерссию вручную на основе тестовой последовательности по средствам аппарата LinearRegression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_test, y_test)
LRx = np.linspace(-5, 20, 100)
LRy = model.coef_*LRx + model.intercept_
LRy = np.transpose(LRy)
plt.plot(LRx, LRy)
plt.plot(x_test, y_test, 'g.')
plt.plot(x_train, y_train, '.')
plt.title('График зависимости Direct_Bilirubin и Total_Bilirubin')
plt.xlabel('Direct_Bilirubin')
plt.ylabel('Total_Bilirubin')
plt.legend(['Линейная регрессия', 'Тестовая выборка', 'Тренировочная выборка'])
```

Посчитать критерий t для двух выборок: больные и здоровые (по критерию Total_Bilirubin). В данном случае сравниваем p-value. Имеют ли данные группы схожесть?

```python
seak = data[data['Dataset'] == 1]['Total_Bilirubin']
healthy = data[data['Dataset'] == 2]['Total_Bilirubin']
stats.ttest_ind(seak, healthy)
```

Уровень p-value значительно ниже, чем допустимое (0.05), поэтому не принимаем нулевую гипотезу, то есть выборки статистически не совпадают.

```python
print(seak.mean())
print(healthy.mean())
sns.distplot(seak, 20)
sns.distplot(healthy, 20)
```

Посчитать статистическую значимость для **трех** выборок (по возрасту) -**тест Anova** - 1ая группа от 1 до 30 лет, 2ая группа от 30 лет до 55, 3я группа более 55 по выборке Dataset. Сравнить с реальностью (они логично что будут отличаться).

```python
low = data[data['Age'] < 30]['Dataset']
mid = data[(data['Age'] >= 30) & (data['Age'] <= 55)]['Dataset']
top = data[data['Age'] > 55]['Dataset']
stats.f_oneway(low, mid, top)
```

P-value значительно меньше заданного значения (0.05), что говорит о статистической значимости и что отвергаем нулевую гипотезу о статистической сходности выборок. То есть 3 выборки статистически не схожи.
